```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import mplcyberpunk

plt.style.use("cyberpunk")

# Generate synthetic exam scores
np.random.seed(10)  # For reproducible results
scores_class_A = np.random.normal(70, 10, 100)  # Class A with mean 70 and std 10
scores_class_B = np.random.normal(80, 15, 100)  # Class B with mean 80 and std 15

# Combine the data
data = np.concatenate([scores_class_A, scores_class_B])
classes = ['Class A'] * 100 + ['Class B'] * 100  # Labels for the classes

# Create a DataFrame for Seaborn
df = pd.DataFrame({'Score': data, 'Class': classes})

# Set up the matplotlib figure
fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Box Plot
sns.boxplot(x='Class', y='Score', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Box Plot')

# Strip Plot
sns.stripplot(x='Class', y='Score', data=df, jitter=True, ax=axes[0, 1])
axes[0, 1].set_title('Strip Plot')

# Swarm Plot
sns.swarmplot(x='Class', y='Score', data=df, ax=axes[0, 2])
axes[0, 2].set_title('Swarm Plot')

# Bar Plot
sns.barplot(x='Class', y='Score', data=df, ax=axes[0, 3])
axes[0, 3].set_title('Bar Plot')

# Point Plot
sns.pointplot(x='Class', y='Score', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Point Plot')

# Histogram
sns.histplot(data=df, x="Score", hue="Class", element="step", stat="density", common_norm=False, ax=axes[1, 1])
axes[1, 1].set_title('Histogram')

# KDE Plot
sns.kdeplot(data=df, x="Score", hue="Class", common_norm=False, ax=axes[1, 2])
axes[1, 2].set_title('KDE Plot')

# Violin Plot
sns.violinplot(x='Class', y='Score', data=df, ax=axes[1, 3])
axes[1, 3].set_title('Violin Plot')

plt.suptitle("Distribution Plots", fontsize=32, y=1.03)

# Adjust layout
plt.tight_layout()
plt.show()
```
